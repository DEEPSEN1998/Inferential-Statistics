# ** THEORY**

1. **What is hypothesis testing in statistics**
- Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating two competing hypotheses, collecting data, and determining whether the evidence supports one hypothesis over the other.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis**
- Null Hypothesis (H₀): This is the default assumption that there is no effect or no difference.
- Alternative Hypothesis (H₁ or Ha): This is the claim that there is a significant effect or difference. It challenges the null hypothesis.

3.  **What is the significance level in hypothesis testing, and why is it important**

- The significance level (α) is the threshold for rejecting the null hypothesis, typically set at 0.05 (5%). It represents the probability of committing a Type I error (false positive). A lower α means stricter criteria for rejecting H₀, reducing false positives but increasing false negatives.

4. **What does a P-value represent in hypothesis testing**
-     The P-value measures the probability of obtaining a test result as extreme as (or more extreme than) the observed data, assuming the null hypothesis is true.
-    Interpretation:
        If P-value ≤ α: Reject the null hypothesis (strong evidence against H₀).
        If P-value > α: Fail to reject the null hypothesis (insufficient evidence).

- Example: If P = 0.03 and α = 0.05, we reject H₀, meaning there is significant evidence in favor of the alternative hypothesis.

5.  **How do you interpret the P-value in hypothesis testing**
- Interpretation: If P-value ≤ α: Reject the null hypothesis (strong evidence against H₀). If P-value > α: Fail to reject the null hypothesis (insufficient evidence).

6. **What are Type 1 and Type 2 errors in hypothesis testing**
- Type I Error (False Positive): Rejecting a true null hypothesis. (Example: Concluding a drug works when it doesn’t.)
- Type II Error (False Negative): Failing to reject a false null hypothesis. (Example: Concluding a drug doesn’t work when it actually does.)

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing**
-     One-Tailed Test: Tests for an effect in one direction (e.g., "greater than" or "less than").
-    Two-Tailed Test: Tests for an effect in both directions (e.g., "different from" without specifying direction).

- Example: Testing whether a new medicine increases recovery speed:

    - One-tailed: Only checks if it’s faster.
    - Two-tailed: Checks if it's either faster or slower.

8. **What is the Z-test, and when is it used in hypothesis testing**
- A Z-test is used to compare sample and population means when the population standard deviation (σ) is known, and the sample size is large (n ≥ 30). It assumes a normal distribution.

   - Example: Testing if the average IQ of students in a school differs from the national average.

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing**
- The Z-score tells how many standard deviations a data point is from the mean. It is calculated as:
Z=(X−μ)/σ
​

- where X = sample mean, μ = population mean, σ = standard deviation.

    - Z > 0: Above the mean
    - Z < 0: Below the mean
    - Higher |Z| values → More extreme results

10.  **What is the T-distribution, and when should it be used instead of the normal distribution**
- The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It has thicker tails, accounting for more variability in small samples.

11. **What is the difference between a Z-test and a T-test**
- Z-Test vs. T-Test

    - Z-Test: Used when σ is known and n ≥ 30.
    - T-Test: Used when σ is unknown and n < 30.

12. **What is the T-test, and how is it used in hypothesis testing**
- A T-test compares means between two groups. Common types:

    - One-sample T-test: Compares a sample mean to a known population mean.
    - Independent (Two-sample) T-test: Compares means of two independent groups.
    - Paired T-test: Compares means before and after treatment within the same group.

13. **What is the relationship between Z-test and T-test in hypothesis testing**
- Both tests compare means, but T-tests handle small sample sizes with unknown standard deviation, while Z-tests require a known population standard deviation. As sample size increases, the T-distribution approaches the Z-distribution.

14. **What is a confidence interval, and how is it used to interpret statistical results**
- A confidence interval (CI) provides a range of values within which the true population parameter is likely to fall. It is calculated as:
CI= X ± Zα/2 *  σ/root of n
- Example: A 95% CI of (50, 60) means we are 95% confident that the true mean lies within this range.

15. **What is the margin of error, and how does it affect the confidence interval**
- The margin of error (ME) quantifies the uncertainty in an estimate. A smaller ME leads to a narrower CI, indicating more precision. ME is influenced by:

    - Sample size (larger n → smaller ME)
    - Variability in data (higher σ → larger ME)
    - Confidence level (higher confidence → larger ME)

16. **How is Bayes' Theorem used in statistics, and what is its significance**
- Bayes' Theorem updates probabilities based on new evidence. It is used in probabilistic reasoning, machine learning, and spam filtering.
  - Formula:
P(A/B)=P(B/A)*P(A)/P(B)
​



17. **What is the Chi-square distribution, and when is it used**
- The Chi-square distribution is used for categorical data analysis. It is applied in tests like:

    - Chi-square goodness of fit test
    - Chi-square test for independence

18. **What is the Chi-square goodness of fit test, and how is it applied**
- Used to test whether an observed frequency distribution matches an expected distribution.

- Example: Checking if dice rolls follow a uniform distribution.

19. **What is the F-distribution, and when is it used in hypothesis testing**
- The F-distribution is used to compare variances in statistical tests, including ANOVA and F-tests. It is skewed right and depends on degrees of freedom.

20. **What is an ANOVA test, and what are its assumptions**
- Analysis of Variance (ANOVA) compares means across three or more groups.

    - Assumptions:
        - Data is normally distributed.
        - Groups have equal variances.
        - Observations are independent.

- Example: Testing if three different teaching methods lead to different exam scores.

21. **What are the different types of ANOVA tests**
- Types of ANOVA Tests

    - One-way ANOVA: Compares means of one factor (e.g., different diets).
    - Two-way ANOVA: Compares means of two factors (e.g., diet and exercise).
    - Repeated measures ANOVA: Used for repeated observations on the same subjects.

22. **What is the F-test, and how does it relate to hypothesis testing?**
- An F-test compares two variances to determine if they are significantly different. It is often used in ANOVA and regression analysis.

- Example: Testing if two manufacturing processes produce outputs with the same variance.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest
from scipy.stats import ttest_ind, ttest_rel
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency, f_oneway, chi2, ttest_ind
```

```python
# 1. Perform a Z-test for comparing a sample mean to a known population mean
def perform_z_test(sample, pop_mean, pop_std, alpha=0.05):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    print(f"Z-score: {z_score}, P-value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
```

```python
# 2. Simulate random data and calculate the P-value
np.random.seed(42)
sample_data = np.random.normal(50, 10, 30)  # Sample from N(50, 10)
perform_z_test(sample_data, pop_mean=55, pop_std=10)
```

```python
# 3. One-sample Z-test
def one_sample_z_test(sample, pop_mean):
    z_stat, p_val = ztest(sample, value=pop_mean)
    print(f"Z-statistic: {z_stat}, P-value: {p_val}")
    return z_stat, p_val

one_sample_z_test(sample_data, 55)
```

```python
# 4. Two-tailed Z-test with visualization
def plot_z_test(sample, pop_mean, pop_std, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.plot(x, y, label='Standard Normal Curve')
    critical_value = stats.norm.ppf(1 - alpha / 2)
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Region')
    plt.axvline(-critical_value, color='red', linestyle='--')

    z_stat, p_val = one_sample_z_test(sample, pop_mean)
    plt.axvline(z_stat, color='blue', linestyle='-', label=f'Z-score: {z_stat:.2f}')

    plt.legend()
    plt.title("Two-Tailed Z-test Decision Region")
    plt.show()

plot_z_test(sample_data, 55, 10)
```

```python
# 5. Type 1 and Type 2 Error Visualization
def type1_type2_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)
    plt.plot(x, y, label='H0 Distribution')
    plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.5, label='Type 1 Error')
    plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.5)
    plt.legend()
    plt.title("Type 1 and Type 2 Errors Visualization")
    plt.show()

type1_type2_errors()
```

```python
# 6. Independent T-test
def independent_t_test(data1, data2):
    t_stat, p_val = ttest_ind(data1, data2)
    print(f"T-statistic: {t_stat}, P-value: {p_val}")

data1 = np.random.normal(50, 10, 30)
data2 = np.random.normal(55, 10, 30)
independent_t_test(data1, data2)
```

```python
# 7. Paired T-test with visualization
def paired_t_test(data1, data2):
    t_stat, p_val = ttest_rel(data1, data2)
    print(f"Paired T-test: T-statistic = {t_stat}, P-value = {p_val}")
    plt.hist(data1 - data2, bins=10, alpha=0.6, color='g', label='Difference Histogram')
    plt.axvline(np.mean(data1 - data2), color='red', linestyle='--', label='Mean Difference')
    plt.legend()
    plt.title("Paired T-test: Differences in Samples")
    plt.show()

paired_t_test(data1, data2)
```

```python
# 8. Compare Z-test and T-test on simulated data
def compare_z_t_tests():
    sample = np.random.normal(50, 10, 20)
    print("Z-test Results:")
    one_sample_z_test(sample, 55)
    print("T-test Results:")
    t_stat, p_val = stats.ttest_1samp(sample, 55)
    print(f"T-statistic: {t_stat}, P-value: {p_val}")

compare_z_t_tests()
```

```python
# 9. Confidence Interval Calculation
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2., len(sample)-1) * sem
    lower, upper = mean - margin, mean + margin
    print(f"{confidence * 100}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return lower, upper

confidence_interval(sample_data)
```

```python
#10. Calculate Margin of Error
def margin_of_error(sample, confidence=0.95):
    sem = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2., len(sample)-1) * sem
    print(f"Margin of Error: {margin:.2f}")
    return margin

# Example Usage
sample = np.random.normal(50, 10, 30)
margin_of_error(sample)
```

```python
# 11. Bayesian Inference
def bayes_theorem(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    print(f"Posterior Probability: {posterior:.2f}")
    return posterior

# Example Usage
bayes_theorem(0.5, 0.8, 0.6)
```

```python
# 14. Goodness-of-Fit Test
from scipy import stats
observed = np.array([30, 14, 34, 45])
expected = np.array([25, 25, 25, 25])

# Scale expected values to match the sum of observed values
expected = expected * (observed.sum() / expected.sum())

chi2_stat, p = stats.chisquare(observed, expected)
print(f"Goodness-of-Fit Chi2 Statistic: {chi2_stat}, P-value: {p}")
```

```python
# 16. F-test for comparing variances
data1 = np.random.normal(50, 10, 30)
data2 = np.random.normal(55, 15, 30)
f_stat = np.var(data1, ddof=1) / np.var(data2, ddof=1)
p_val = 1 - stats.f.cdf(f_stat, len(data1)-1, len(data2)-1)
print(f"F-statistic: {f_stat}, P-value: {p_val}")
```

```python
# 25. Z-test for Comparing Proportions
def z_test_proportions(p1, p2, n1, n2):
    prop = (p1 * n1 + p2 * n2) / (n1 + n2)
    z_score = (p1 - p2) / np.sqrt(prop * (1 - prop) * (1/n1 + 1/n2))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    print(f"Z-score: {z_score}, P-value: {p_value}")

z_test_proportions(0.6, 0.5, 100, 100)
```
